````python
# Функция для построения гистограмм
def plot_histograms(dataframe, title):
    # Отбираем только числовые столбцы, кроме 'id'
    numeric_columns = dataframe.select_dtypes(include=['float64', 'int64']).drop(columns='id', errors='ignore')
    
    # Построение гистограмм
    numeric_columns.hist(bins=20, figsize=(15, 10), edgecolor='black')
    plt.suptitle(title, fontsize=16)
    plt.show()

# Построение гистограмм для каждой таблицы
plot_histograms(market_file, "Распределение количественных признаков (market_file)")
plot_histograms(market_money, "Распределение количественных признаков (market_money)")
plot_histograms(market_time, "Распределение количественных признаков (market_time)")
plot_histograms(money, "Распределение количественных признаков (money)")
````

## Корреляционный анализ

Не все признаки распределены нормально, поэтому используем корреляцию Phik.

````python
# Выбор количественных признаков для анализа
quantitative_features = [
    'маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность', 
    'акционные_покупки', 'средний_просмотр_категорий_за_визит', 
    'неоплаченные_продукты_штук_квартал', 'ошибка_сервиса', 'страниц_за_визит',
    'выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц', 
    'время_предыдущий_месяц', 'время_текущий_месяц'
]

# Рассчитываем PhiK корреляцию между количественными признаками
phik_corr = merged_data.phik_matrix(interval_cols=quantitative_features)

# Визуализация корреляционной матрицы PhiK
plt.figure(figsize=(12, 10))
sns.heatmap(phik_corr, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"shrink": 0.8})
plt.title("PhiK Корреляционная матрица")
plt.show()

# Фильтрация корреляций, исключая NaN и вывод только корреляций больше 0.9
high_corr = phik_corr.stack().reset_index()
high_corr.columns = ['feature_1', 'feature_2', 'phi_k_value']

# Оставляем только те строки, где корреляция больше 0.9, исключая саморефлексию (где feature_1 == feature_2)
high_corr = high_corr[(high_corr['phi_k_value'] > 0.9) & (high_corr['feature_1'] != high_corr['feature_2'])]

# Выводим результаты
print("Пары признаков с высокой корреляцией (больше 0.9):")
print(high_corr)
````

В данных отсутствуют пары признаков с высокой корреляцией (выше 0.9), это может означать, что мультиколлинеарность не является серьёзной проблемой.
   Мультиколлинеарность возникает, когда два или более признаков сильно коррелируют друг с другом, что может привести к нестабильности моделей, например, к переоценке значимости переменных. В вашем случае не выявлены сильные корреляции (выше 0.9), при этом присутствуют **умеренные** корреляции, которые могут негативно повлиять на модель.

## Пайплайны

**Выбор метрики для задачи:**

Бизнес-задача предполагает акцент на удержании постоянных клиентов.
- **Ошибкой первого рода** (precision) будет ситуация, когда модель ошибочно предсказала, что активность клиента снизилась, хотя на самом деле она осталась на прежнем уровне.
- **Ошибка второго рода** (recall) - это ситуация, когда модель предсказывает, что клиент останется на прежнем уровне, хотя на самом деле его активность снижается.

Для бизнеса важно не пропускать клиентов со снижающейся активностью, чтобы вовремя среагировать и не упустить их. Это важнее, чем ошибочно сделать выгодное предложение клиентам с прежней активностью, при этом, ошибку первого рода также будем учитывать.

В нашем случае, закодируем целевую переменную "покупательская_активность" как бинарную, где:
0 - "Прежний уровень" и 1 - "Снизилась") и будем использовать метрику F2-Score, в которой в два раза больше важности придается recall по сравнению с precision.

### ColumnTransformer

**Подготовка пайплайна для предобработки данных:**
1. **Обработка количественных признаков**: Используем два скейлера: `StandardScaler` и `MinMaxScaler`, чтобы проверить, как они влияют на модели.
2. **Обработка категориальных признаков**: Используем два кодировщика: `OneHotEncoder` и `OrdinalEncoder`, чтобы проверить, как каждый из них влияет на обучение моделей.

Создадим пайплайн.

````python
# Пайплайн для обработки категориальных признаков с OrdinalEncoder
ord_pipe = Pipeline([
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),  # Заполнение пропусков
    ('encoder', OrdinalEncoder(categories=[
#         ['стандарт', 'премиум'],  # тип_сервиса
#         ['нет', 'да'],  # разрешить_сообщать
        ['не_определено', 'меньше_половины', 'почти_все'],  # акционная_группа
        ['низкая', 'средняя', 'высокая']  # маркет_актив_тек_мес_категория
    ]))  
])

# Пайплайн для обработки категориальных признаков с OneHotEncoder
ohe_pipe = Pipeline([
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),  # Заполнение пропусков
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))  # OHE
])

# Пайплайн для обработки количественных признаков с масштабированием
num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Заполнение пропусков
    ('scaler', StandardScaler())  # Default StandardScaler, который будет заменен в RandomizedSearchCV
])

# Комбинированный ColumnTransformer для всех типов признаков
preprocessor = ColumnTransformer([
    ('ord', ord_pipe, ord_columns),
    ('ohe', ohe_pipe, ohe_columns),
    ('num', num_pipe, num_columns)
])
````

Зададим модели и параметры для RandomizedSearchCV

````python
# Гиперпараметры для поиска
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    
    # словарь для модели KNeighborsClassifier()
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1')],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    
    # словарь для модели SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'models__C': [0.1, 1, 10],
        'models__kernel': ['linear', 'rbf'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
````

Примененим RandomizedSearchCV с новым словарем.

````python
# Уберём лишние предупреждения, не влияющие на результат
warnings.filterwarnings("ignore", category=FutureWarning)

# Итоговый пайплайн с препроцессором и моделью
pipe_final = Pipeline([
    ('preprocessor', preprocessor),
    ('models', LogisticRegression(random_state=RANDOM_STATE))  # По умолчанию LogisticRegression
])

# Подбор гиперпараметров с RandomizedSearchCV
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    scoring=f2_scorer,  # Используем f2_scorer для оценки качества
    random_state=RANDOM_STATE,
    verbose=1,
    n_jobs=-1
)

# Замер времени
start_time = time.time()

# Обучаем модель
randomized_search.fit(X_train, y_train)

# Время завершения обучения
end_time = time.time()

# Вычисляем время, потраченное на обучение модели
execution_time = end_time - start_time

# Лучшие параметры модели
display('Лучшая модель и её параметры:\n', randomized_search.best_estimator_)
print('Метрика на обучающей выборке:', randomized_search.best_score_)

# Оценка на тестовой выборке
y_test_pred = randomized_search.predict(X_test)

f2_test = fbeta_score(y_test, y_test_pred, beta=2)
print(f'F2-Score на тестовой выборке: {f2_test}')
# print(f'ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_test_pred)}')
# print(f'Accuracy на тестовой выборке: {accuracy_score(y_test, y_test_pred)}')

# Выводим время, затраченное на выполнение пайплайна
print(f'Время, затраченное на выполнение пайплайна: {execution_time:.2f} секунд')
````

**5. Применение для маркетинга**

````python
# Порог для вероятности снижения
probability_threshold = 0.8

# Порог для прибыли
profitability_threshold = 2

# Применяем пороги для создания сегмента
final_data['segment'] = 'Other'  # По умолчанию 'Other'
final_data.loc[
    (final_data['predict_proba'] >= probability_threshold) & 
    (final_data['прибыль'] >= profitability_threshold), 
    'segment'
] = 'High Risk, High Profit'

# Выведем результаты сегментации
high_risk_profit = final_data[final_data['segment'] == 'High Risk, High Profit']

# display(high_risk_profit)
````

3. **Интерактивные элементы на платформе:** Напоминания о незавершенных покупках, рекомендации товаров на основе предыдущих покупок, улучшенная навигация.

6. **Анализ неоплаченных продуктов:** Рекомендуется проанализировать причины неоплаты продуктов и предложить клиентам, попавшим в этот сегмент, специальные условия оплаты, рассрочку или скидки для повышения вероятности завершения покупок.

7. **Контент и рекомендации:** Для группы "почти все", которая активно посещает страницы, рекомендуется использовать перекрестные продажи (cross-sell) и предлагать дополнительные товары, которые могут заинтересовать клиента.

8. **Анализ времени на сайте:** Повышение вовлеченности клиентов, которые проводят больше времени на сайте, может быть достигнуто через персонализированные предложения, эксклюзивный контент и целевые акции.
